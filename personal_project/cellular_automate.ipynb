```python
import numpy as np
import matplotlib.pyplot as plt

from jupyterthemes import jtplot
jtplot.style(context='notebook', fscale=1.2, spines=False, gridlines='--')
```

```python
datafiles = ['live_test.log', 'coral_test.log', 'strange_damn.log']
err = -0.0000000000001

with open(datafiles[0]) as f:
    f = f.readlines()

live_cells = []
dead_cells = []
time_deltas = []
t_sum = 0

for line in f:
    elems = line.split('/')
    live_cells.append(int(elems[0]))
    dead_cells.append(int(elems[1]))
    time_deltas.append(float(elems[2]))
    t_sum += float(elems[2])

time_deltas = np.array(time_deltas)
t = np.arange(0, t_sum+err, time_deltas.mean())

plt.plot(t, live_cells, label='Живые клетки')
plt.plot(t, dead_cells, label='Мертвые клетки')
plt.legend(loc='best')
```
