```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
```

```python
# начальные условия
a = 0.77
b = 0.33
c = 0.077
d = 0.033

x0 = [4, 9]
t = np.arange(0, 400, 0.1)
```

```python
def syst(x, t):
    dx1 = -a*x[0] + c*x[0]*x[1]
    dx2 = b*x[1] - d*x[0]*x[1]
    return [dx1, dx2]
```

```python
y = odeint(syst, x0, t)
y
```

```python
y1 = y[:, 0]
y2 = y[:, 1]
```

```python
# построение графика колебаний изменения числа популяции хищников
plt.title('Популяция хищников')
plt.plot(t, y1)
```

```python
# построение графика колебаний изменения числа популяции жертв 
plt.title('Популяция жертв')
plt.plot(t, y2)
```

```python
# построение графика зависимости изменения численности хищников от изменения численности жертв
plt.title('Зависимость численности хищников от численности жертв')
plt.plot(y1, y2)
```
